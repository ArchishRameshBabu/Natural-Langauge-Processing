```python
##Identifying the start and end of conversation based on double quotes 
start_position=[]
end_position=[]

for i in range(0,len(Raw_Data)):
    if Raw_Data.loc[Raw_Data.index[i],'POS'] =='``':
        start_position.append(i)
    if Raw_Data.loc[Raw_Data.index[i],'POS'] ==Raw_Data.loc[211,'POS']:
        end_position.append(i)
    
```

```python
##Collating results 
positions=sorted(start_position+end_position)
positions=pd.DataFrame(positions)
positions.columns=['start']
positions['end']=positions['start'].shift(-1)
positions_fn=positions.iloc[::2]
positions_fn=positions_fn.values
positions_fn=pd.DataFrame(positions_fn)
positions_fn=positions_fn.astype(int)
```

```python
##Collating results 
x=[]
for i in range(positions_fn.shape[0]):
    st=positions_fn.loc[i][0]
    ed=positions_fn.loc[i][1]+3
    x.append(list(Raw_Data.loc[st:ed]['word']))
```

The output below gives the word position and the text corresponding to identified combination, using these text values a person can tag the speaker and listerners. 

In the first row we see that the 205th word is where the coversation start and it ends at word position 211. Also from the text in the next column we see that the speaker is Mr.White(he is the main character in the book)
