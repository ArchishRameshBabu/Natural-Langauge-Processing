```python
import pandas as pd
import numpy as np
#import nltk
#import math

##Reading the text
Raw_Data = pd.read_csv("C:/Users/archi/OneDrive/Desktop/ANLP Assignment/tokens_v4.csv",encoding = "ISO-8859-1")
```

```python
##Printing the data
Raw_Data.head()
```

```python
##Creating empty arrays
sentance_Position=[]
Char_Position = []
speaker=[]
```

```python
##Identifying the speaker and the sentance 
for i in range(0,len(Raw_Data)):
    if Raw_Data.loc[Raw_Data.index[i],'NER'] =='PERSON':
        if Raw_Data.loc[Raw_Data.index[i-1],'word'] =='Mr.' or Raw_Data.loc[Raw_Data.index[i-1],'word'] =='Mrs.':
            Char_Position.append(Raw_Data.loc[Raw_Data.index[i], 'Row_nbr'])
            speaker.append(Raw_Data.loc[Raw_Data.index[i-1], 'word'] + Raw_Data.loc[Raw_Data.index[i], 'word'])
            sentance_Position.append(Raw_Data.loc[Raw_Data.index[i], 'Sentence_Nbr'])
        else: 
            Char_Position.append(Raw_Data.loc[Raw_Data.index[i], 'Row_nbr'])
            speaker.append(Raw_Data.loc[Raw_Data.index[i], 'word'])
            sentance_Position.append(Raw_Data.loc[Raw_Data.index[i], 'Sentence_Nbr'])
   
```

```python
##Collating and formatting results
Char_Position=pd.DataFrame(Char_Position)
sentance_Position=pd.DataFrame(sentance_Position)
speaker=pd.DataFrame(speaker)
df=pd.concat([Char_Position,speaker],axis=1)
df1=pd.concat([df,speaker],axis=1)
df1.columns = ['char start position', 'Speaker','Sentance']
```

```python
##Accounting for consequtive conversations between the same set of people 
count=1
Final_DF=[]
for i in range(1,df1.shape[0]):
    if df1.loc[df1.index[i-1],'Speaker']==df1.loc[df1.index[i],'Speaker']:
       count+=1
    else :
        Final_DF.append([df1.iloc[i-1]['Speaker'],count])
        count=1
```

The below table gives the speaker, the other person in the coversation and the number of times they spoke during each conversation. The results are in the same order as they appear in the book. If x and y spoke during seperate instances then each instance is tagged seperately. Multiple conversations which happen during the same instance are aggreated in the Conversations column. 

The code also incorrectly tags the author, publisher as part of the charecters

```python
Final_DF=pd.DataFrame(Final_DF)
Final_DF.columns = ['Speaker','Count of times they spoke']
Final_DF['Other Speaker'] = Final_DF['Speaker'].shift(-1)
print(Final_DF)
```
